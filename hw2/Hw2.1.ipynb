```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
```

```python
#Замініть у таблиці значення "—" на значення NaN
df4 = pd.read_html(url, match="Коефіцієнт народжуваності в регіонах України", decimal=',', thousands=" ", na_values="—")
df4[0]
```

```python
#Визначте типи всіх стовпців за допомогою dataframe.dtypes
df5 = df[0]
df5.dtypes
```

```python
#Замініть типи нечислових колонок на числові
df6 = df4[0]
df6['2014'] = df6['2014'].fillna(0)
df6['2019'] = df6['2019'].fillna(0)
df6
```

```python
#Порахуйте, яка частка пропусків міститься в кожній колонці
df7 = df4[0]
none = (df7.isnull().sum() / len(df7)) * 100
print(none)
```

```python
#Видаліть з таблиці дані по всій країні, останній рядок таблиці
df8 = df4[0]
df8 = df8.drop(df8.index[-1])
df8
```

```python
#Замініть відсутні дані в стовпцях середніми значеннями цих стовпців
df9 = df4[0]
df9 = df9.fillna(df9.mean(numeric_only=True))
df9
```

```python
#Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
df10 = df4[0]
df_selected = df10[['Регіон', '2019']].copy()
df_selected.loc['2019'] = pd.to_numeric(df_selected['2019'], errors='coerce')
average_2019 = df_selected['2019'].mean()
result = df_selected[df_selected['2019'] > average_2019]
result[['Регіон', '2019']]
```

```python
#У якому регіоні була найвища народжуваність у 2014 році?
df11 = df4[0]
df_selected = df11[[('Регіон'), ('2014')]].copy()
df_selected[('2014')] = pd.to_numeric(df_selected[('2014')], errors='coerce')
result = df_selected.loc[df_selected[('2014')].idxmax()]
result
```

```python
#Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році
df12 = df4[0]
df_selected = df12[['Регіон', '2019']].copy()
df_selected['2019'] = df_selected['2019'].fillna(0)
df_selected['2019'] = pd.to_numeric(df_selected['2019'], errors='coerce')
df_selected = df_selected.sort_values(by='2019', ascending=False)
#-----
plt.figure(figsize=(12, 6))
plt.bar(df_selected['Регіон'], df_selected['2019'], color='skyblue')
plt.xlabel('Регіон')
plt.ylabel('Рівень народжуваності в 2019 році')
plt.title('Рівень народжуваності по регіонам України в 2019 році')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
