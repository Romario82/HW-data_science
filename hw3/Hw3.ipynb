```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import random
```

```python
#Функція гіпотези лінійної регресії у векторному вигляді
def mean_squared_error(y_true, y_pred):
  mse = np.mean((y_true - y_pred) ** 2)
  return mse
```

```python
#Функція для обчислення функції втрат у векторному вигляді
def cross_entropy_loss(y_true, y_pred):
  y_pred = np.clip(y_pred)
  ce_loss = np.sum(y_true * np.log(y_pred))
  return ce_loss
```

```python
#Один крок градієнтного спуску
def one_step_gradient(X, y, weights, lr = 0.001):
    m = len(y)
    predictions = np.dot(X, weights)
    errors = predictions - y
    gradient = (1/m) * np.dot(X.T, errors)
    w = weights - lr * gradient
    return w
```

```python
data_fr2 = data_fr[["area", "bedrooms", "bathrooms"]]
target = data_fr['price']
data_fr2.head()
```

```python
m = np.array(data_fr2)
m
```

```python
matr_X = data_fr2.values
print(len(matr_X))
```

```python
matr_y = np.array(target)
print(len(matr_y))
```

```python
#weights = weights - lr * np.dot(x.transpose(), (y_pred - y)) / m
w = np.array([random. uniform(-5, 5) for i in range(matr_X.shape[1] + 1)])
w
```

```python
matr = np.ones((len(matr_X), 1))
X = np.concatenate((matr, matr_X), axis=1)
result = one_step_gradient(X, matr_y, w)
result
```

```python
model = LinearRegression()
regress = model.fit(matr_X, matr_y)
print(regress.coef_)
print(regress.intercept_)
y_test = model.predict(matr_X)
print(y_test)
print(f"Середньо квадратична помилка - {mean_squared_error(matr_y, y_test)}")
print(regress.predict(np.array([[7500, 4, 2]])))
```
